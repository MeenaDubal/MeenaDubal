#What is EDA?

Exploratory Data Analysis: this is unavoidable and one of the major step to fine-tune the given data set(s) in a different form of analysis to understand the insights of the key characteristics of various entities of the data set like column(s), row(s) by applying Pandas, NumPy, Statistical Methods, and Data visualization packages.

#Out Come of this phase as below

Understanding the given dataset and helps clean up the given dataset. It gives you a clear picture of the features and the relationships between them.

Providing guidelines for essential variables and leaving behind/removing non-essential variables.

Handling Missing values or human error.

Identifying outliers

EDA process would be maximizing insights of a dataset.

### One Sample Test For Mean 
let's Suppose, it was observed that the average purchase made by the Men of the age 18-25 was 10000. Is it still the same? 

```python
#T test - it is used to compare means of two samples
#Assumption: Sample size < 30
# pop variance unknown

#Null hyp : Both means are equal
#alt hyp : Both means are not equal


#one sample t test : we compare the mean of sample with some standard value
```

```python
#Null hy : mean is = 10000
#Alt : mean not equal to 10000
```

```python
new_data = data.loc[(data["Age"] == 1) & (data["Gender"] == 1)]
new_data
```

```python
new_data.shape
```

```python
sample_size = 29
sample = new_data.sample(sample_size)
```

```python
sample
```

```python
pos_mean = 10000
```

```python
sample_mean = sample['Purchase'].mean()
print(sample_mean)
```

```python
#one sample t test
from scipy.stats import ttest_1samp
t_stat, p_value = ttest_1samp(sample["Purchase"],pos_mean)
print(t_stat,p_value)
```

```python
#p is less than 0.05 so we should reject our null hyp
```

#F test--- compares the variance of two samples



```python
data_men = data.loc[(data["Gender"] == 1) & data["Age"] == 1]
data_women = data.loc[(data["Gender"] == 0) & data["Age"] == 1]
```

```python
#Creating samples
data_men_sample = data_men.sample(500)
data_women_sample = data_women.sample(500)

```

```python
from dataclasses import dataclass
#checking variance
print(data_men_sample.Purchase.var())
print(data_women_sample.Purchase.var())
```

```python
print(data_men_sample.Purchase.mean())
print(data_women_sample.Purchase.mean())
```

```python
from scipy.stats import f
F = data_men_sample.Purchase.mean()/data_women_sample.Purchase.mean()
F
```

```python
#Degree of freedom
df1 = len(data_men_sample) - 1
df2 = len(data_women_sample) - 1

print(df1, df2)
```

```python
#for p value
import scipy
scipy.stats.f.cdf(F,df1,df2)
```

```python
#p value > 0.05 so accept the null hyp
```
